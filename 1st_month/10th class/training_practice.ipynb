# Training

```python
import pandas as pd
from datacleaner import autoclean
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder,  MinMaxScaler,  RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('movies.csv')
df.head(3)
```

```python
df['Gross'].value_counts()
```

```python
df.drop(['ONE-LINE', 'GENRE'], axis=1, inplace=True)
df = autoclean(df)
df.head(3)
```

```python

df.dtypes
```

```python
x = df.drop(['Gross'], axis = 1)
y = df['Gross']
```

```python
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.2,random_state=42)
x_test,x_val,y_test,y_val=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
```

```python
x_train
```

```python
x_val
```

```python
log_model=LogisticRegression(random_state=42)
```

```python
log_model.fit(x_train,y_train)
```

```python
log_pred=log_model.predict(x_val)
```

```python
log_pred[3]
```

```python
log_score=accuracy_score(y_val,log_pred)
```

```python
print('Accuracy:',log_score)
```

```python
log_pred_test=log_model.predict(x_test)
```

```python
log_score_test = accuracy_score(y_test,log_pred_test)
print(log_score_test)
```

```python
log_pred_test
```

```python
from sklearn.model_selection import cross_val_score,KFold
import numpy as np
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(log_model, x, y, cv=kf, scoring='neg_mean_squared_error')
cv_scores = np.sqrt(-cv_scores)

print('Cross-Validation-Scores:', cv_scores)
print("Mean Cross Validation Score:", np.mean(cv_scores))
```
