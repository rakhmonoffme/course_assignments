# Encoding

```python
import pandas as pd
```

```python

df = pd.read_csv('telefon turini aniqlash2.csv')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Fill missing values
# For categorical columns, fill with the mode
# For numerical columns, fill with the mean
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].mean())

df.isnull().sum()
```

```python
df['Rangi'] = df['Rangi'].str.replace(' ', '')
```

```python
df['Rangi']
```

```python
rangi_dummies = pd.get_dummies(df['Rangi'], prefix='Rangi', drop_first=True, dtype='int')
df = pd.concat([df, rangi_dummies], axis=1)
df.drop('Rangi', axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
df['Rangi'] = encoder.fit_transform(df['Rangi'])
```

```python
df['Rangi']
```

```python
df.head()
```

```python
df['Rangi'].nunique() # find unique values
```
